```python
# Types of categorical plots:
# factorplot, boxplot, violinplot, stripplot, swarmplot, barplot, countplot
import seaborn as sb
import numpy as np
%matplotlib inline
tips = sb.load_dataset('tips')
tips.head()
```

```python
# Barplot and countplot
# Allow you to get aggregate data off a categorical feature in your data.
# Barplot allows you to aggregate by some function (such as mean)
sb.barplot(x="sex", y="total_bill", data=tips)
```

```python
sb.barplot(x="sex", y="total_bill", data=tips, estimator=np.std)
```

```python
# Countplot is essentially the same as barplot, but the estimator is explicitly counting the number of occurrences
# Hence, why only the x value is passed in
sb.countplot(x="sex", data=tips)
```

```python
# Boxplots and Violinplots
# Used to show the distribution of categorical data
# Box plot shows the distribution of quantitative data in a way that facilitates comparisons between variables or across levels
# of a categorical variable.
# The box shows the quartiles of the dataset while the whiskers extend to show the rest of the distribution (except points that
# are determined to be outliers using a method that is a function of the inter-quartile range)
sb.boxplot(x="day", y="total_bill", data=tips, palette="pastel")
```

```python
# Can set orientation (i.e. horizontal)
# Can display entire data frame
sb.boxplot(data=tips, palette="pastel", orient="h")
```

```python
# In this way, a provided x value must be numeric
# This won't work: sb.boxplot(x="day", y="total_bill", data=tips, orient="h")
sb.boxplot(x="total_bill", y="size", data=tips, orient="h")
```

```python
sb.boxplot(x="day", y="total_bill", hue="smoker", data=tips, palette="pastel")
```

```python
# Violinplot
# Unlike boxplot, the violinplot features a kde of the underlying distribution
sb.violinplot(x="day", y="total_bill", data=tips, palette="pastel")
```

```python
sb.violinplot(x="day", y="total_bill", hue="sex", data=tips, palette="Set2")
```

```python
sb.violinplot(x="day", y="total_bill", hue="sex", data=tips, split=True, palette="Set2")
```

```python
# Stripplot
# Draws a scatterplot where one variable is categorical.
# Good complement to a box or violin plot in cases where you want to show all observations along with some representation
# of the underlying distribution
sb.stripplot(x="day", y="total_bill", data=tips)
```

```python
sb.stripplot(x="day", y="total_bill", data=tips, jitter=False)
```

```python
sb.stripplot(x="day", y="total_bill", hue="sex", data=tips, palette="Set2")
```

```python
sb.stripplot(x="day", y="total_bill", hue="sex", data=tips, dodge=True, palette="Set2")
```

```python
# Swarmplot
# similar to stripplot, but the data points are adjusted so they don't overlap (does not scale well with large amounts of points)
sb.swarmplot(x="day", y="total_bill", data=tips)
```

```python
sb.swarmplot(x="day", y="total_bill", hue="sex", data=tips, palette="Set2", dodge=True)
```

```python
# COMBINING CATEGORICAL PLOTS
sb.violinplot(x="tip", y="day", data=tips, palette="pastel")
sb.swarmplot(x="tip", y="day", data=tips, color="black", size=3)
```

```python
# Factorplot (now renamed catplot)
# Most general form of a categorical plot. Takes a kind parameter to adjust plot type
sb.catplot(x="sex", y="total_bill", data=tips, kind="bar")
```

```python
sb.catplot(x="sex", y="total_bill", data=tips, kind="strip")
```
