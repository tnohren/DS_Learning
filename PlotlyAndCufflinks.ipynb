```python
import pandas as pd
import numpy as np
import cufflinks as cf
from plotly import __version__
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
%matplotlib inline

print(__version__)
```

```python
init_notebook_mode(connected=True)
```

```python
cf.go_offline()
```

```python
df = pd.DataFrame(np.random.randn(100, 4), columns='A B C D'.split())
df.head(5)
```

```python
df2 = pd.DataFrame({'Category':['A', 'B', 'C'], 'Values': [32, 43, 50]})
df2.head()
```

```python
df.iplot(kind="scatter", x='A', y='B', mode='markers', size=10)
```

```python
df2.iplot(kind='bar', x='Category', y='Values', theme='ggplot')
```

```python
df.count().iplot(kind='bar', theme='ggplot')
```

```python
df.iplot(kind='box')
```

```python
df3 = pd.DataFrame({'x': [1, 2, 3, 4, 5], 'y': [10, 20, 30, 40, 50], 'z': [5, 4, 3, 2, 1]})
df3.iplot(kind="surface", colorscale="reds")
```

```python
df[['A', 'B']].iplot(kind='spread')
```

```python
df['A'].iplot(kind='hist', bins=20)
```

```python
df.iplot(kind='bubble', x='A', y='B', size='C')
```
