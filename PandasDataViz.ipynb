```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

df1 = pd.read_csv('df1', index_col=0)
df2 = pd.read_csv('df2')
df1.head()
```

```python
# Matplotlib includes stylesheets to make plots look nicer
# Includes plot_bmh, plot_fivethirtyeight, plot_ggplot, etc.
```

```python
# Before setting style
df1['A'].hist()
```

```python
# After setting style
plt.style.use('ggplot')
df1['A'].hist()
```

```python
# Several built in plot types for pandas
# df.plot.area, df.plot.barh, df.plot.density, df.plot.hist, df.plot.line,
# df.plot.scatter, df.plot.bar, df.plot.box, df.plot.hexbin, df.plot.kde,
# df.plot.pie
# Can also use df.plot(kind='hist') or replace hist with any of the other types
```

```python
# Area
df2.plot.area(alpha=0.4)
```

```python
df2.plot(kind='area', alpha=0.4)
```

```python
# Barplots
df2.plot.bar()
```

```python
df2.plot.bar(stacked=True)
```

```python
# Histograms
df1['A'].plot.hist(bins=50, edgecolor='white')
```

```python
# Line Plot
df1.plot.line(x='A', y='B', figsize=(12, 3), lw=1)
```

```python
# Scatter Plot
df1.plot.scatter(x='A', y='B')
```

```python
# Use c parameter to apply color based on another column value
df1.plot.scatter(x='A', y='B', c='C', cmap='coolwarm')
```

```python
# Use s parameter to apply size based on another column value. Must be array
df1.plot.scatter(x='A', y='B', s=df1['C']*200, edgecolors='white')
```

```python
# Box Plots
df2.plot.box() # by= argument for groupby
```

```python
# Hexagonal Bin Plot
# Useful for bivariate data - alternative to scatterplot
tempdf = pd.DataFrame(np.random.randn(1000, 2), columns=['A', 'B'])
tempdf.plot.hexbin(x='A', y='B', gridsize=25, cmap='coolwarm')
```

```python
# KDE Plots
df2['a'].plot.kde(lw=5)
```

```python
df2.plot.density()
```
