```python
import matplotlib.pyplot as plt
import numpy as np

# For jupyter notebooks only
%matplotlib inline

# Ranges for x axis and y axis
x = np.linspace(0, 5, 11)
y = x**2
```

```python
plt.plot(x, y, 'b')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Plot Title')
plt.show()
```

```python
plt.subplot(1, 2, 1) # subplot(numrows, numcols, plotnumber)
plt.plot(x, y, 'b--')
plt.subplot(1, 2, 2)
plt.plot(y, x, 'g*-')
plt.show()
```

```python
# Object-Oriented Method
fig = plt.figure()

# Add Axes - left, bottom, width, height
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

# Plot
axes.plot(x, y, 'b')
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')

# Show Plot
plt.show()
```

```python
# Create blank canvas
fig = plt.figure()

# Create Axes
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # Larger Axis
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) # Smaller Axis

# Larger Plot
axes1.plot(x, y, 'b')
axes1.set_xlabel('X Axis 1')
axes1.set_ylabel('Y Axis 1')
axes1.set_title('Title 1')

# Smaller Plot
axes2.plot(y, x, 'g')
axes2.set_xlabel('X Axis 2')
axes2.set_ylabel('Y Axis 2')
axes2.set_title('Title 2')

# Show plot
plt.show()
```

```python
# plt.subplots() object = more automatic axis manager
fig, axes = plt.subplots()
axes.plot(x, y, 'b')
axes.set_xlabel('X Axis')
axes.set_ylabel('Y Axis')
axes.set_title('Title')
plt.show()
```

```python
# Empty plot with 1x2 subplots
fig, axes = plt.subplots(nrows=1, ncols=2)
plt.show()
```

```python
# Iterate through axes
for axis in axes:
    axis.plot(x, y, 'b')
    axis.set_xlabel('X Axis')
    axis.set_ylabel('Y Axis')
    axis.set_title('Title')
    
fig
```

```python
# Solve overlapping subplots
fig, axes = plt.subplots(nrows=1, ncols=2)

for axis in axes:
    axis.plot(x, y, 'b')
    axis.set_xlabel('X Axis')
    axis.set_ylabel('Y Axis')
    axis.set_title('Title')
    
fig
plt.tight_layout()
```

```python
fig, axes = plt.subplots(figsize=(8,4), dpi=100)

axes.plot(x, y, 'r')
axes.set_xlabel('X Axis')
axes.set_ylabel('Y Axis')
axes.set_title('Title')

plt.show()
```

```python
# Legends
fig = plt.figure()
axis = fig.add_axes([0, 0, 1, 1])
axis.plot(x, x**2, 'b.-', label="x^2")
axis.plot(x, x**3, 'r--', label="x^3")
axis.legend(loc=2) # loc=0 (let mpl decide), 1 (upper right), 2 (upper left), 3 (lower left), 4 (lower right)
plt.show()
```

```python
# Line and Marker Styles
fig, axis = plt.subplots(figsize=(12, 6))

# linestype options: "-", "--", "-.", ":", "steps"
# marker options: "+", "o", "*", "s", ",", ".", "1", "2", "3", "4", ...
axis.plot(x, x+1, color="red", linewidth=0.25)
axis.plot(x, x+2, color="red", linewidth=0.50)
axis.plot(x, x+3, color="red", linewidth=1.00)
axis.plot(x, x+4, color="red", linewidth=2.00)

axis.plot(x, x+5, color="green", lw=3, linestyle="-")
axis.plot(x, x+6, color="green", lw=3, ls="-.")
axis.plot(x, x+7, color="green", lw=3, ls=":")

line, = axis.plot(x, x+8, color="black", lw=1.5)
line.set_dashes([5, 10, 15, 20])

axis.plot(x, x+9, color="blue", lw=3, ls="-", marker="+")
axis.plot(x, x+10, color="blue", lw=3, ls="--", marker="o")
axis.plot(x, x+11, color="blue", lw=3, ls="-", marker="s")
axis.plot(x, x+12, color="blue", lw=3, ls="--", marker="1")

axis.plot(x, x+13, color="purple", lw=1, ls="-", marker="o", markersize=2)
axis.plot(x, x+14, color="purple", lw=1, ls="-", marker="o", markersize=4)
axis.plot(x, x+15, color="purple", lw=1, ls="-", marker="o", markersize=8,
         markerfacecolor="red")
axis.plot(x, x+16, color="purple", lw=1, ls="-", marker="s", markersize=8,
         markerfacecolor="yellow", markeredgewidth=3, markeredgecolor="green")
```

```python
fig, axes = plt.subplots(1, 3, figsize=(12, 4))

for i in range(0, len(axes)):
    axes[i].plot(x, x**2, x**3)
    axes[i].set_title("Default Axes Ranges" if i==0 else ("Tight Axes" if i==1 else "Custom Axes Ranges"))
    axes[i].set_ylim([0, 60])
    axes[i].set_xlim([2, 5])
    if i==1: axes[i].axis('tight')
        
plt.show()
```
