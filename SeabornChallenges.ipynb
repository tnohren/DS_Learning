```python
import seaborn as sb
import matplotlib.pyplot as plt
%matplotlib inline

sb.set_style("whitegrid")
titanic = sb.load_dataset('titanic')
titanic.head(5)
```

```python
sb.jointplot(data=titanic, x="fare", y="age", kind="scatter")
```

```python
sb.distplot(titanic['fare'], kde=False, color='red')
```

```python
sb.boxplot(data=titanic, x='class', y='age')
```

```python
sb.swarmplot(data=titanic, x='class', y='age', palette='pastel', dodge=True)
```

```python
sb.countplot(data=titanic, x='sex')
```

```python
sb.heatmap(data=titanic.corr(), cmap="coolwarm")
plt.title('Pairwise Correlation')
```
