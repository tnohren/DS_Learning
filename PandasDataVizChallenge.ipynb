```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

df3 = pd.read_csv('df3')
df3.head()
```

```python
df3.plot.scatter(x='a', y='b', color='red', edgecolors='black', s=50, figsize=[12, 3])
```

```python
df3['a'].plot.hist(edgecolor='black', color='blue')
```

```python
plt.style.use('ggplot')
df3['a'].plot.hist(edgecolor='white', alpha=0.5, bins=25)
```

```python
df3[['a', 'b']].plot.box()
```

```python
df3['d'].plot.kde()
```

```python
df3['d'].plot.kde(lw=5, ls='--')
```

```python
df3.iloc[0:30].plot.area(alpha=0.4)
```

```python
fig = plt.figure()
df3.iloc[0:30].plot.area(alpha=0.4, ax=fig.gca())
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```
