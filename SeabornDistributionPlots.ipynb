```python
# Plots that allow distribution of a data set:
# distplot, jointplot, pairplot, rugplot, kdeplot
import seaborn as sb
%matplotlib inline

# tips is a seaborn built-in dataset
tips = sb.load_dataset('tips')
tips.head()
```

```python
# DISTPLOTS
# Shows the distribution of a univariate set of observations
sb.distplot(tips['total_bill'])
```

```python
# Removes kernel density estimation and uses 10 hist plot bins
sb.displot(tips['total_bill'], kde=False, bins=10)
```

```python
# JOINTPLOT
# Match up two distplots for bivariate data. With choice of what kind parameter to compare with:
# scatter, reg, resid, kde, hex
sb.jointplot(x='total_bill', y='tip', data=tips, kind='scatter')
```

```python
sb.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
```

```python
sb.jointplot(x='total_bill', y='tip', data=tips, kind='reg')
```

```python
sb.jointplot(x='total_bill', y='tip', data=tips, kind='resid')
```

```python
sb.jointplot(x='total_bill', y='tip', data=tips, kind='kde')
```
