```python
import seaborn as sb
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Grids are general types of plots that allow you to map plot types to rows and columns of a grid.
# This helps create similar plots separated by features
iris = sb.load_dataset('iris')
iris.head()
```

```python
# Pairgrid
# subplot grid for plotting pairwise relationships in a dataset
sb.PairGrid(iris)
```

```python
pg = sb.PairGrid(iris)
pg.map(plt.scatter)
```

```python
# Map to upper, lower, and diagonal
pg = sb.PairGrid(iris)
pg.map_diag(plt.hist) # Histogram
pg.map_upper(plt.scatter) # Scatterplot
pg.map_lower(sb.kdeplot) # Kernel Density Estimation
```

```python
# Pairplot
# simpler version of pairgrid
sb.pairplot(iris)
```

```python
pp = sb.pairplot(data=iris, hue="species", palette="pastel")
```

```python
# Facetgrid
# General way to create grids of plots based off of a feature
tips = sb.load_dataset('tips')
tips.head()
```

```python
fg = sb.FacetGrid(data=tips, col="time", row="smoker")
```

```python
fg = fg.map(plt.hist, "total_bill")
```

```python
fg = sb.FacetGrid(data=tips, col="time", row="smoker", hue="sex")
fg = fg.map(plt.scatter, "total_bill", "tip").add_legend()
```
