```python
# Will cover regression in much later detail later
# only cover lmplot() for now
# lmplot allows you to display linear models, but also allows you to split up those plots based off of features, as well
# as the coloring the hue based off of features.
import seaborn as sb
import matplotlib.pyplot as plt
%matplotlib inline

tips = sb.load_dataset('tips')
sb.lmplot(data=tips, x="total_bill", y="tip")
```

```python
sb.lmplot(data=tips, x="total_bill", y="tip", hue="sex", palette="pastel")
```

```python
# WORKING WITH MARKERS
# lmplot kwargs get passed through to regplot which is a more general form of lmplot().
# regplot has a scatter_kws parameter that gets passed to plt.scatter
# Want to set 's' parameter in that dictionary, which corresponds to squared marker size.
# http://matplotlib.org/api/markers_api.html
sb.lmplot(data=tips, x="total_bill", y="tip", hue="sex", palette="pastel",
             markers=['$m$', '$f$'], scatter_kws={'s': 100})
```

```python
# USING A GRID
# We can add more variable separation through columns and rows with the use of a grid.
# Indicate with col or row arguments
sb.lmplot(data=tips, x="total_bill", y="tip", col="sex", palette="pastel")
```

```python
sb.lmplot(data=tips, x="total_bill", y="tip", row="sex", col="time")
```

```python
# Can adjust size and aspect ratio
sb.lmplot(data=tips, x="total_bill", y="tip", hue="sex", col="day", palette="pastel", aspect=0.6, height=10)
```
