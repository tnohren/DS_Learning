```python
import pandas as pd
import numpy as np
from numpy.random import randn
np.random.seed(101)

# Initialize Test Data Frame
df = pd.DataFrame(randn(5, 4), index="A B C D E".split(), columns="W X Y Z".split())
df
```

```python
# Access Multiple Columns of Data Frame
df[['W', 'Z']]
```

```python
# Creating New Column
df['WY'] = df['W'] + df['Y']
df
```

```python
# Dropping Column
df.drop('WY', axis=1, inplace=True)
```

```python
df
```

```python
# Dropping Row
df.drop('E', axis=0) # not using inplace, so data kept
```

```python
# Selecting Rows off Label
df.loc['A']
```

```python
# Selecting Rows off Position
df.iloc[0]
```

```python
# Selecting Subset of Rows and Columns
df.loc['B', 'Y']
```

```python
# Wider Range
df.loc[['A', 'B'], ['W', 'Y']]
```

```python
# Conditional Selection
cond = df>0
cond
```

```python
df[cond]
```

```python
# Selecting all where W column > 0
df[df['W']>0]
```

```python
# Result is data frame, so we can apply data frame operations
df[df['W']>0][['Y', 'X']]
```

```python
# For and, must use &
df[(df['W']>0) & (df['X']>0)]
```

```python
# For or, must use |
df[(df['W']>0) | (df['X']>0)]
```

```python
# Reset Index
df.reset_index()
```

```python
newind = 'CA NY WY OR CO'.split()
df['States'] = newind
df.set_index('States', inplace=True)
df
```

```python
# Multi-indexing and index hierarchy
outside = ['G1', 'G1', 'G1', 'G2', 'G2', 'G2']
inside = [1, 2, 3, 1, 2, 3]
hier_index = list(zip(outside, inside)) # Creates Tuple of Index Pairs
hier_index = pd.MultiIndex.from_tuples(hier_index)
hier_index
```

```python
df = pd.DataFrame(np.random.randn(6, 2), index=hier_index, columns=['A', 'B'])
df
```

```python
df.loc['G1']
```

```python
df.loc['G1'].loc[1]
```

```python
# Change index names
df.index.names = ['Group', 'Num']
df
```

```python
df.xs('G1')
```

```python
df.xs(('G1', 1))
```

```python
df.xs(1,level='Num')
```
